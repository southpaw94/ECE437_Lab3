```python
import numpy as np
from numpy.fft import fft, ifft
import matplotlib.pyplot as plt
from scipy.io import loadmat
from scipy.signal import periodogram

%matplotlib inline
plt.rcParams['figure.figsize'] = (10, 8)

%load_ext autoreload
%autoreload 2
```

```python
lab3_audio = loadmat('lab3Audio.mat')
lab3_audio = lab3_audio['lab3Audio']

Fs = 22050
Ts = 1 / Fs
```

```python
time = np.arange(lab3_audio.shape[0])
x = lab3_audio

plt.plot(time, x)
plt.grid()
plt.title('Audio signal, x(t)')
plt.show()
```

```python
n = len(lab3_audio)
Px = np.sum(x ** 2) / n
Xf = fft(x) / Fs
Exf = np.abs(Xf) ** 2
Pxf = Exf / (n * Ts)

freq = np.arange(0, n, 1) * Fs / n
maxPxf = np.max(Pxf)

plt.plot(freq[:int(n/2)], Pxf[:int(n/2)] * 100)
plt.xlim(0, Fs / 2)
plt.ylim(0, maxPxf)
plt.title('Power spectral density of x(t)')
plt.show()
```

```python
f, Px = periodogram(x.flatten(), Fs, nfft=512, return_onesided=True)
plt.plot(f, Px)
plt.title('PSD of noisy audio in dB')
plt.xlabel('Frequecy (Hz)')
plt.ylabel('Gain (dB)')
plt.grid()
plt.show()
```

```python
Xk = fft(x)
Xsize = len(Xk)
Hk = np.concatenate([np.ones(10000),
                     np.zeros(10000),
                     np.ones(10000)])

plt.plot(freq[:int(n / 2)], Hk[:int(n / 2)])
plt.xlim(0, Fs / 2)
plt.ylim(-0.2, 1.2)
plt.title('Low pass filter chosen for filtering out noise')
plt.xlabel('Frequency (Hz)')
plt.ylabel('H(f)')
plt.show()
```

```python
Xkfilter = Xk.flatten() * Hk
xfilter = ifft(Xkfilter)

f, Px = periodogram(np.real(xfilter), Fs, 
                    nfft=512, return_onesided=True)

plt.plot(f, Px, lw=2)
plt.title('PSD of filtered audio')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Gain (dB)')
plt.grid()
plt.show()
```

```python
samples = np.arange(Xsize)
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
ax = ax.flatten()

ax[0].plot(samples, np.real(x), lw=2)
ax[0].set_title('Noisy signal')
ax[1].plot(samples, np.abs(Xk), lw=2)
ax[1].set_title('Spectrum of noisy signal')
ax[2].plot(samples, np.real(xfilter), lw=2)
ax[2].set_title('Filtered signal')
ax[3].plot(samples, np.real(Xkfilter), lw=2)
ax[3].set_title('Spectrum of clean signal')

plt.show()
```
